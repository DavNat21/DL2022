## Exercise 8 - Polynomial fits with TensorFlow

Repeat the previous exercise by replacing the Numpy primitives with TensorFlow primitives.


```python
def true_function(x):
    return np.cos(1.5*np.pi*x)

def mse(y, ymodel):
    
    squared_error = (y-y_model)**2
    loss = squared_error / y.size
    
    return loss

x = np.sort(np.random.rand(30))
y_sample = true_function(x) + np.random.rand()*0.1
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y_sample)
plt.scatter(x, true_function(x), label='True')
plt.title('True_function(x) vs random noise')
plt.legend()
plt.show()
```

To implement the polynomial fit, i'll use the **np.polyfit** method.  

```python
coeff1 = np.polyfit(x,y_sample, deg=1)
coeff2 = np.polyfit(x,y_sample, deg=4)
coeff3 = np.polyfit(x,y_sample, deg=15)

y_fit1 = np.polyval(coeff1,x)
y_fit2 = np.polyval(coeff2,x)
y_fit3 = np.polyval(coeff3,x)
```

```python
figure, ax = plt.subplots(1,3, figsize=(20,5))
ax[0].plot(x,y_fit1)
ax[1].plot(x,y_fit2)
ax[2].plot(x,y_fit3)
for j in range(3):
    ax[j].plot(x,y_sample)
ax[0].set_title('poly deg 1')
ax[1].set_title('poly deg 4')
ax[2].set_title('poly deg 15')
```
